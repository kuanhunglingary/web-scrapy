```python
# Artccle 取得單一新聞
from newspaper import Article
import json
# 自由時報
url = 'https://news.ltn.com.tw/news/politics/breakingnews/4560654'
article=Article(url)
article.download()
print(article.html)
article.parse()
article.title
# parse() 解析
article.parse()
print('新聞標題')
print(article.title)
print('新聞內容：')
print(article.text)
print('新聞日期')
print(article.publish_date)
news_data = {
    "新聞標題": article.title,
    "新聞內容": article.text,
    "新聞日期": str(article.publish_date)
}

json_data = json.dumps(news_data, ensure_ascii=False, indent=2)
with open("politics_news.json", "w", encoding="utf-8") as json_file:
    json_file.write(json_data)
print("News data saved to politics_news.json")
```

```python
import requests as req
from bs4 import BeautifulSoup as bs
from pprint import pprint
url = "https://news.ltn.com.tw/list/breakingnews/politics"

# 用 requests 的 get 方法把網頁抓下來
res = req.get(url) 

# 指定 html.parse 作為解析器
soup = bs(res.text, "html.parser")
# 透過迭代取得所有 a 的屬性 href
for a in soup.select('a'):
    if a.has_attr('href'):
        print(a['href']) # a.get("href")
    else:
        print("=" * 50)
        print(f"連結[{a.get_text()}] 沒有 href 屬性")
        print("=" * 50)
```
